```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import tensorflow as ts
```

```python
def create_plot_with_borders(data, title, cmap, levels, extend, cbar_label):
    # Use cartopy and create a PlateCarree projection that is centered at 180 degrees (b/c hemispheres are flip) 
    projection = ccrs.PlateCarree(central_longitude=180)
    fig, ax = plt.subplots(figsize=(12,7), subplot_kw={'projection': projection})
    ax.add_feature(cfeature.BORDERS, linestyle='-', alpha=.5)
    ax.coastlines(resolution='110m')
    # Contour plot 
    cs = ax.contourf(data.lon, data.lat, data, levels=levels, cmap=cmap, extend=extend, transform=ccrs.PlateCarree())
    # Addinng color bar
    cbar = fig.colorbar(cs, orientation='vertical', fraction=0.02, pad=0.04)
    ax.set_title(title)
    # Setting global extent 
    ax.set_extent([-180, 180, -90, 90], crs=projection)
    # Gridlines and labels 
    ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
    # Setting the aspsect of the plot to be equal 
    ax.set_aspect('auto', adjustable=None)
    return fig, ax
```

```python
# Load the NetCDF file
file_path = '/Users/maliawalewski/Downloads/research/datasets_airTemp/ssp1-2.6_airTemp.nc'
data = xr.open_dataset(file_path)

# Define the parameters for the plot
histclim_title = 'CMIP6 ENSMN ssp1_2.6 future climate (Air Temp) (2020-2049)'
histclim_cmap = 'jet'
histclim_levels = np.arange(-40, 41, 4) 
histclim_extend = 'both' 
histclim_cbar_label = 'Near-Surface Air Temperature (shading)/Sea-Ice Area Percentage (Atmospheric Grid) (contour) ANN'

# Use the historical climate data for plotting
fig_histclim, ax_histclim = create_plot_with_borders(data['histclim'], histclim_title, histclim_cmap, histclim_levels, histclim_extend, histclim_cbar_label)
```
